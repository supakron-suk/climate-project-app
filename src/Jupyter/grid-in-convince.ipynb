```python
import geopandas as gpd
import matplotlib.pyplot as plt

#  GeoJSON
data = gpd.read_file('../Geo-data/nc_to_json_2000_1.json')
shapefile = gpd.read_file('../Geo-data/shapefile-lv1-thailand.json')

#  (  "Nakhon Ratchasima")
korat_coord = shapefile[shapefile['NAME_1'] == 'Nakhon Ratchasima']

#  data 
# 
grid_in_korat = data[data.geometry.intersects(korat_coord.geometry.unary_union)]

# 
fig, ax = plt.subplots(figsize=(10, 10))

# Plot 
korat_coord.plot(ax=ax, color='white', edgecolor='black', alpha=1, label='Nakhon Ratchasima')

# Plot 
grid_in_korat.plot(ax=ax, color='red', edgecolor='black', alpha=0.5, label='Grid in Korat')

# 
for idx, grid in grid_in_korat.iterrows():
    # 
    grid_area = grid.geometry.area
    
    # 
    intersection_area = grid.geometry.intersection(korat_coord.geometry.unary_union).area
    
    # 
    intersection_percentage = (intersection_area / grid_area) * 100
    
    #  temperature 
    temperature_value = grid['temperature']
    
    
    # 
    # 
    x, y = grid.geometry.centroid.x, grid.geometry.centroid.y
    ax.text(x, y, f'{intersection_percentage:.2f}%\nTemp: {temperature_value:.2f}', fontsize=8, ha='center', color='black')
    print(f"Grid {idx}: intersects {intersection_percentage:.2f}%, Temp: {temperature_value:.2f}")
    

# 

#print(grid_in_korat)
plt.title('Grid in Nakhon Ratchasima', fontsize=16)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

#  (legend)
plt.legend()

# 
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

#  GeoJSON
data = gpd.read_file('../Geo-data/nc_to_json_2001.json')
shapefile = gpd.read_file('../Geo-data/thailand-Geo.json')

#  (  "Bangkok Metropolis")
province_name = 'Nakhon Ratchasima'
province_coord = shapefile[shapefile['NAME_1'] == province_name]

#  data 
# 
grid_in_province = data[data.geometry.intersects(province_coord.geometry.union_all())]

# 
province_area = province_coord.geometry.union_all().area

# 
fig, ax = plt.subplots(figsize=(10, 10))

# Plot 
province_coord.plot(ax=ax, color='white', edgecolor='black', alpha=1, label=province_name)

# Plot 
grid_in_province.plot(ax=ax, color='red', edgecolor='black', alpha=0.5, label=f'Grid in {province_name}')

# 
total_weighted_temp = 0
total_percentage = 0

for idx, grid in grid_in_province.iterrows():
    # 
    intersection_area = grid.geometry.intersection(province_coord.geometry.union_all()).area
    
    # 
    intersection_percentage_of_province = (intersection_area / province_area) * 100
    
    #  temperature 
    temperature_value = grid['temperature']
    
    temperature_value = np.nan_to_num(temperature_value, nan=0.0)
    
    # 
    weighted_temp = temperature_value * intersection_percentage_of_province
    total_weighted_temp += weighted_temp # Zigma
    print(f"{temperature_value} *{intersection_percentage_of_province} = {weighted_temp}")
    total_percentage += intersection_percentage_of_province # Zigma
    print(f"____________________________________")
    print(f"--------total per = {total_percentage}-----")
    
    # 
    # 
    x, y = grid.geometry.centroid.x, grid.geometry.centroid.y
    ax.text(x, y, f'{intersection_percentage_of_province:.2f}%\nTemp: {temperature_value:.2f}', fontsize=8, ha='center', color='black')
    print(f"Grid {idx}: covers {intersection_percentage_of_province:.3f}% of {province_name}, Temp: {temperature_value:.2f}")

# 
average_temperature = total_weighted_temp / total_percentage #if total_percentage != 0 else None
print(f'{province_name} covers all grid: {total_percentage:.2f}%')
print("------------------------------------------------------------------------")
print(f'Average Temperature in {province_name} : {average_temperature:.2f}')
print("------------------------------------------------------------------------")
# 
plt.title(f'Grid in {province_name}', fontsize=16)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

#  (legend)
plt.legend()

# 
plt.show()

```
